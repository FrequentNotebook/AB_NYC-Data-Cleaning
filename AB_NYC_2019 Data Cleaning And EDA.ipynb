## 2.2. Name Column

### Expectation?

The name column is the listing name for each listing. It should be a text based column.

In section 1, we saw that there is a 0.033% of null values in the columnns.

```python
# Inspect the name column

print(f'{df.columns[1]} has {df[df.columns[1]].isna().sum()} null values')
print(f'{df.columns[1]} has {df[df.columns[1]].duplicated().sum()} duplicate rows')
print(f'{df.columns[1]}: {df[df.columns[1]].dtype}')
```

```python
# filter out the null values as a dataframe

name_null_mask = df['name'].isna()
df.loc[name_null_mask]
```

```python
# check for stale listing

df.loc[name_null_mask, ['name', 'number_of_reviews', 'availability_365', 'price']].sort_values('number_of_reviews')
```

```python
# filter out the stale listings

stale_mask = ((df['number_of_reviews'] == 0) 
              & (df['availability_365'] == 0)
             & df['name'].isna())
    
df[stale_mask].shape
```

```python
# drop the stale listings

df = df[~stale_mask].reset_index(drop=True)
df.shape
```

```python
# fill the null values in name column with 'unknown name'

df['name'] = df['name'].fillna('Unknown')
```

```python
# Validate the fill

df['name'].isna().sum()
```

```python
# filter out the duplicates in the name column

dupes_name_mask = ((df['number_of_reviews'] == 0) &
                   (df.duplicated(subset=['name', 'host_id', 'neighbourhood', 
                                        'room_type'], keep=False))
)

```

```python
# view filter

df.loc[dupes_name_mask, ['name', 'host_id', 'neighbourhood', 
                                        'room_type', 'number_of_reviews']].sort_values('name')
```

```python
# drop filter

df.sort_values('id', inplace=True)

df = df[
    ~dupes_name_mask |
    ~df.duplicated(subset=['name', 'host_id', 'neighbourhood', 'room_type'], keep='first')
].reset_index(drop=True)
```

```python
print(f'{df.columns[1]} has {df[df.columns[1]].isna().sum()} null values')
print(f'{df.columns[1]} has {df[df.columns[1]].duplicated().sum()} duplicate rows')
print(f'{df.columns[1]}: {df[df.columns[1]].dtype}')
df.shape
```

## Name Column Cleaning

### Null Values
The `name` column had 16 missing values. Upon inspection, 8 of these listings had zero reviews and zero availability — indicating stale/abandoned listings with no transaction history and no active availability on the platform. These 8 rows were dropped.

The remaining 8 null listings had either reviews or availability, suggesting they were real active listings. Since the `name` column is descriptive and not used in analysis, nulls were filled with `'Unknown'` to preserve these rows.

### Duplicates
Duplicates were defined as rows sharing the same `name`, `host_id`, `neighbourhood`, and `room_type` — with `number_of_reviews == 0` on all occurrences. Listings with reviews were excluded from this definition as they represent real transaction history and are considered legitimate separate listings on the platform.

The `id` column was used as a tiebreaker — the lowest `id` (original listing) was kept and all subsequent duplicates were dropped.

**Duplicates before cleaning:** 989     
**Duplicates after cleaning:** 877     
**Duplicates dropped:** 112

**Rows before cleaning:** 48,858  
**Rows after cleaning:** 48,783  
**Rows dropped:** 75

## 2.4. Host_Name Column

### Expectation?
`host_name` is the display name for hosts on the platform. Duplicates are expected and normal for this column since hosts can bear the same name and a single host can have multiple listings.   
In our basic inspection, we can expect to see 21 null values.

```python
# Inspect the host_name column

print(f'{df.columns[3]} has {df[df.columns[3]].isna().sum()} null values')
print(f'{df.columns[3]} has {df[df.columns[3]].duplicated().sum()} duplicate rows')
print(f'{df.columns[3]}: {df[df.columns[3]].dtype}')
```

```python
# Filter out the null value from the main dataframe

host_name_mask = df['host_name'].isna()
df.loc[host_name_mask]
```

```python
# Check for stale listing

host_name_stale = ((df['host_name'].isna()) &
                  (df['number_of_reviews'] == 0) &
                  (df['availability_365'] == 0))
df.loc[host_name_stale]
```

```python
# drop the stale listing

df = df[~host_name_stale].reset_index(drop=True)
df.shape
```

```python
# map host_name to host_id from existing records and fill null host_names

host_name_to_id = df.dropna(subset=['host_name']).groupby('host_id')['host_name'].first()
df['host_name'] = df['host_name'].fillna(df['host_id'].map(host_name_to_id))
```

```python
# check if we still have missing values

df['host_name'].isna().sum()
```

```python
# Fill the null values in host_name witj 'unknown'

df['host_name'] = df['host_name'].fillna('unknown')
```

```python
# validate the fill

df['host_name'].isna().sum()
```

```python
# Validate the host_name column

print(f'{df.columns[3]} has {df[df.columns[3]].isna().sum()} null values')
print(f'{df.columns[3]} has {df[df.columns[3]].duplicated().sum()} duplicate rows')
print(f'{df.columns[3]}: {df[df.columns[3]].dtype}')
```

## Host_Name Column Cleaning

After inspecting the null values in the `host_name` column, we realized that there we some stale listings. These were listings that do not have a `host_name`, no `number_of_reviews`, and no `availability_365`. We dropped the stale listings.    

We went on to try to map host_name to host_id from the existing records. This helps to check if there is a host_name for the same host_id in our existing record. Unfortunately, there wasn't.
